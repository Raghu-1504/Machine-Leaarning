```python
X_train = train.drop('ACTION',axis = 1)
y_train = train['ACTION']
```

```python
X_train
```

# Feature scaling

```python
from sklearn.preprocessing import RobustScaler


rrs =RobustScaler()
X_train= rrs.fit_transform(X_train)
X_test = rrs.transform(X_test)
```

```python
train
```

```python
from catboost import CatBoostClassifier
cat = CatBoostClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(cat,scoring = 'accuracy',cv = 5,param_grid  = {'max_depth' : [3,8],"n_estimators": [500, 700],"learning_rate": [0,0.015]},verbose=False,n_jobs = -1)

gs.fit(X_train, y_train)

print("Best Parameters:", gs.best_params_)
print("Best Score:", gs.best_score_)
```

# model fitting using bestparameters from the grid search cv

```python
import catboost as ca
cat = ca.CatBoostClassifier(max_depth =8 ,n_estimators =700 ,learning_rate =.015)
cat.fit(X_train,y_train)
```

```python
train_predictions = cat.predict(X_train)
test_predictions = cat.predict(X_test)
X_train_predictions = pd.DataFrame(train_predictions)
X_test_predictions = pd.DataFrame(test_predictions)
```

```python
X_train_predictions.value_counts(),y_train.value_counts()
```

```python
X_test_predictions.value_counts(),y_test.value_counts()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5)) 
sns.histplot(X_train_predictions, bins=2, ax=axes[0])
axes[0].set_title( 'Train_predicted')
sns.histplot(y_train, bins=2, ax=axes[1])
axes[1].set_title(' Orginal_values')
               
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, test_predictions)
print("Accuracy:", accuracy)
```

```python
prediction = cat.predict(X_test)
print(cat.score(X_test,y_test)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(prediction,y_test))
```
