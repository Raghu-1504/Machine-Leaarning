```python
df.drop('User ID',axis = 1,inplace = True)
```

```python
df.head()
```

# **Data SplitÂ¶

```python
y = df['Purchased']
y
```

```python
X = df.drop('Purchased',axis = 1)
X
```

# Split test and train

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 2)
```

# Label transformation

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train['Gender'] = le.fit_transform(X_train['Gender'])
X_test['Gender'] = le.fit_transform(X_test['Gender'])
```

```python
X_train
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

#from sklearn.preprocessing import Normalizer
 
ssc = StandardScaler()
X_train= ssc.fit_transform(X_train)
X_test = ssc.transform(X_test)
```

```python
X_train
```

# using K - Fould CrossValidation

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression 
log = LogisticRegression()
score = cs = cross_val_score(log,X_train, y_train,cv = 5, scoring='accuracy')
print("Cross-validation scores:", score*100)
np.std(score)*100,np.mean(score)*100
```

# MODEL Training

```python
log.fit(X_train, y_train)
print(log.score(X_train,y_train)*100)
```

```python
log.predict_proba(X_test)
```

```python
predictions  = log.predict(X_test)
predictions
```

```python
y_test.values
```

```python
print('test analysis',log.score(X_test,y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test,predictions)
```

```python
print (classification_report(predictions,y_test))
```
