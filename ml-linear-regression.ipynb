```python
df['Date'] = pd.to_datetime(df['Date'],format='mixed')
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
#df['date'] = df['Date'].dt.day
def new(Month):
    if Month in [3,4,5,6]:
        return 'Summer'
    elif Month in [7,8,9,10]:
        return 'rainy'
    else:
        return 'winter'
df['month'] = df['Month'].apply(new)
```

```python
df.drop('Fuel_Price',axis = 1,inplace = True)
```

```python
a = ['Temperature','CPI','Unemployment']
for i in a :
    df[i] = np.log(df[i])
```

```python
X = df.drop('Weekly_Sales',axis = 1)
y = df['Weekly_Sales']
```

```python
X.isnull().sum()
```

```python
X['Temperature'] = X['Temperature'].fillna(method = 'ffill')
```

```python
X = X.dropna(axis = 0)
```

```python
X.isnull().sum()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)
```

```python
X_train
```

```python
from sklearn.preprocessing import RobustScaler
rrs =RobustScaler()
X_train= rrs.fit_transform(X_train)
X_test = rrs.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
ele = ElasticNet()
```

```python
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100],'l1_ratio': [.001,0.1, 0.3, 0.5, 0.7, 0.9]}
```

```python
grid_search = GridSearchCV(estimator= ele,param_grid= param_grid,cv=5,scoring='r2')
grid_search.fit(X_train,y_train)    
best_params = grid_search.best_params_
#best_estimator = grid_search.best_estimator_
#best_score = grid_search.best_score_
print(best_params)
```

```python
ele = ElasticNet(alpha = 0.01, l1_ratio =  .1)
ele.fit(X_train,y_train)
```

```python
ele.score(X_train,y_train)*100
```

```python
pred = ele.predict(X_test)
```

```python
r2_score(y_test,pred)*100
```

```python
from sklearn import metrics
print('MSE :::::',metrics.mean_squared_error(y_test,pred)*100)
```
