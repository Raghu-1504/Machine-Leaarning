```python
sns.kdeplot(df['upvotes'],clip = (0,200))###clip parameter restricts the x-axis range to values between 0 and 200
```

```python
plt.figure(figsize = (6,4))
sns.kdeplot(df.loc[df['featured']==0,'upvotes'],clip =(0,100),color = 'green',label = 'non featured')
sns.kdeplot(df.loc[df['featured']==1,'upvotes'],color = 'red',label = 'featured')
plt.xlim(0,100)
plt.legend(('yes','no'),title = 'Featured ?')






# plt.figure(figsize=(8,5))
# sns.kdeplot(df.loc[df['featured'] == 0, 'upvotes'], color='green', label='non-featured')
#sns.kdeplot(df.loc[df['featured'] == 1, 'upvotes'], color='red', label='featured')
# plt.xlim(0, 100) # Limit the view from 0 to 100
# plt.show()
```
