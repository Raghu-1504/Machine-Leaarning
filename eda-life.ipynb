```python
df = pd.read_csv('/kaggle/input/life-expectancy-and-socio-economic/life expectancy.csv')
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

**from the above we can clearly say that  Corruption feature has most null values so we can drop it **

```python
df.drop('Corruption',inplace = True,axis = 1)
```

```python
df.head(5)
```

```python
df["Undernourishment"] = df['Prevelance of Undernourishment']
df.drop('Prevelance of Undernourishment',axis = 1 ,inplace = True)
```

```python
df
```

```python
null = []
for i in df.columns:
    if df[i].isnull().sum() > 0 :
        null.append(i)
print(null)
```

```python
for i in null:
    df[i] = df[i].fillna(method = 'ffill',)
```
