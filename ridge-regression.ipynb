# FEATURE ENGINEERING

```python
cat_columns = df.select_dtypes(include = 'object').columns
cat_columns
```

```python
df.duplicated().value_counts(normalize = True)*100
```

```python
df.isnull().sum()
```

```python
df  = pd.get_dummies(df,cat_columns,drop_first = True)
```

```python
df.head(4)
```

```python
sns.heatmap(df.corr(),annot = True)
```

```python
df.columns
```

```python
y = df['mpg']
```

```python
X= df[['displacement', 'horsepower', 'weight', 'acceleration',
       'model year', 'cylinders_4', 'cylinders_5', 'cylinders_6',
       'cylinders_8', 'origin_2', 'origin_3']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha = .3)
ridge.fit(X_train,y_train)
```

```python
print(ridge.score(X_train,y_train)*100)
```

```python
predictions = ridge.predict(X_test)
predictions
```

```python
print('train analysis',ridge.score(X_train,y_train)*100)
print('test analysis',ridge.score(X_test,y_test)*100)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, predictions)*100
```
