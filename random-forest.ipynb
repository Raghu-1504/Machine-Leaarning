```python
df  = df.drop_duplicates()
co = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
       'marital_status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week','native-country',
       'income']
df.columns = co
df.columns = df.columns.str.replace('-','_')
```

# Splitting the data into X&Y

```python
X  = df.drop('income',axis = 1)
y = df['income']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=True,test_size  = .3,random_state = 0)
```

# Feature engineering


```python
x = [i for i in X.columns if df[i].dtypes == 'O' ]
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in x:
    X_train[i] = encoder.fit_transform(X_train[i])
    X_test[i]= encoder.transform(X_test[i])
```

```python
X_train.head(3)
```

# Feature scaling

```python
from sklearn.preprocessing import RobustScaler


rrs =RobustScaler()
X_train= rrs.fit_transform(X_train)
X_test = rrs.transform(X_test)
```

# Importing model(Decision tree)

```python
# from sklearn.ensemble import RandomForestClassifier
# rc = RandomForestClassifier(random_state = 0,max_depth=4,)
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=4,random_state = 0)
```

# Using K fold cross validation

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(dt,X_train, y_train,cv = 5, scoring='accuracy')
print("Cross-validation scores:", score*100)
np.std(score)*100,np.mean(score)*100
```

# Fitting the model

```python
dt.fit(X_train,y_train)
print(dt.score(X_train,y_train)*100)
```

```python
prediction = dt.predict(X_test)
print(dt.score(X_test,y_test)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(prediction,y_test))
```

# DT with grid search CV

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=True,test_size  = .25,random_state = 10)

# Feature engineering

x = [i for i in X.columns if df[i].dtypes == 'O' ]
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in x:
    X_train[i] = encoder.fit_transform(X_train[i])
    X_test[i]= encoder.transform(X_test[i])
    
#Feature scaling

from sklearn.preprocessing import RobustScaler


rrs =RobustScaler()
X_train= rrs.fit_transform(X_train)
X_test = rrs.transform(X_test)

# selecting the model (decision tree)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()

```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(dt,scoring = 'accuracy',cv = 10,param_grid  = {'criterion':['gini','entropy'],'max_depth' : range(1,15),
'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4]})

# fit the model

gs.fit(X_train, y_train)

print("Best Parameters:", gs.best_params_)
print("Best Score:", gs.best_score_)
```

```python
print("Best Score:", gs.best_score_*100)
```

```python
dtt = DecisionTreeClassifier(criterion =  'entropy', max_depth=9, min_samples_leaf = 2, min_samples_split= 5)
dtt.fit(X_train,y_train)
print(dtt.score(X_train,y_train)*100)
```

```python
prediction = dtt.predict(X_test)
print(dtt.score(X_test,y_test)*100)
from sklearn.metrics import classification_report
print(classification_report(prediction,y_test))
```
