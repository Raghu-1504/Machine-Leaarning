# TEST & TRAIN SPLIT

```python
df.columns
```

```python
y = df['Yearly Amount Spent']
```

```python
X = df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=11)
```

```python
X_train
```

# training MODEL

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_train
```

```python
lm.coef_
```

```python
print (f'coefficients {lm.coef_}')
```

```python
predictions = lm.predict(X_test)
predictions
```

# Result Analysis

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, predictions)*100
```

# ###FEATURE ENGINEERING# # # 

encoding techniques (one hot encoding)

```python
cat_columns = df.select_dtypes(include = 'object').columns
cat_columns
```

```python
df  = pd.get_dummies(df,cat_columns,drop_first = True)
```

```python
df.head()
```

```python
df.columns
```

```python
y = df['expenses']
```

```python
X = df[['age', 'bmi', 'sex_male', 'children_1', 'children_2',
       'children_3', 'children_4', 'children_5', 'smoker_yes',
       'region_northwest', 'region_southeast', 'region_southwest']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
lm.coef_
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, predictions)*100
```
