```python
X = df.drop('Outcome',axis = 1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state = 0)
```

```python
X_train
```

```python
from sklearn.preprocessing import RobustScaler


rrs =RobustScaler()
X_train= rrs.fit_transform(X_train)
X_test = rrs.transform(X_test)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()

from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(xgb,scoring = 'accuracy',cv = 5,param_grid  = {'max_depth' : [3,5,8],"n_estimators": [500, 600, 700],"learning_rate": [0.01,0.015],
'gamma': [0,0.2],'reg_lambda': [0,1.0]})

# fit the model

gs.fit(X_train, y_train)

print("Best Parameters:", gs.best_params_)
print("Best Score:", gs.best_score_)

#model.fit(X_train, y_train)
```

```python
import xgboost as xg
xgb = xg.XGBClassifier(gamma = 0.2,learning_rate = 0.015,max_depth = 3,n_estimators = 700,reg_lambda = 1.0)
xgb.fit(X_train,y_train)
```

```python
predict = xgb.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predict)*100
print("Accuracy of the Logistic Regression Model: ",accuracy)
```
