# Split the data********

```python
y  = df['class']
X = df.drop('class',axis = 1,)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=True,test_size  = .25,random_state = 0)
```

```python
X_train
```

# # categorical to numerical(ENCODING)

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
x = [i for i in X.columns if df[i].dtypes == 'O' ]
x
```

```python
for i in x:
    X_train[i] = encoder.fit_transform(X_train[i])
    X_test[i]= encoder.transform(X_test[i])
```

```python
X_train
```

# using K-Fold Cross validation on DT

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=3,random_state = 0)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(dt,X_train, y_train,cv = 5, scoring='accuracy')
print("Cross-validation scores:", score*100)
np.std(score)*100,np.mean(score)*100
```

```python
dt.fit(X_train,y_train)
```

```python
print('train scores',dt.score(X_train,y_train)*100)
```

```python
prediction = dt.predict(X_test)
prediction
```

```python
print('test_scores',dt.score(X_test,y_test)*100)
```

```python
pre = pd.Series(prediction)
pre
```

```python
pre.value_counts()
```

```python
from sklearn.metrics import classification_report
print(classification_report(prediction,y_test))
```

```python
plt.figure(figsize = (15,8))
from sklearn import tree
tree.plot_tree(dt.fit(X_train, y_train))
```
