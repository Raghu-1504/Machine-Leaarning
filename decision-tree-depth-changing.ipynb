```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/kaggle/input/car-evaluation/car_evaluation.csv')
c = ['buying','maint','doors','persons','lug_boot','safety','class']
df.columns = c
y  = df['class']
X = df.drop('class',axis = 1,)
```

# Training the model by making the depth as 4

```python
#Splitting test and train data

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=True,test_size  = .25,random_state = 0)

## label encoding for all columns

c = ['buying','maint','doors','persons','lug_boot','safety','class']
df.columns = c
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
x = [i for i in X.columns if df[i].dtypes == 'O' ]
for i in x:
    X_train[i] = encoder.fit_transform(X_train[i])
    X_test[i]= encoder.transform(X_test[i])
```

```python
#Training the model by setting Max depth as 4

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=4,random_state = 0)

# using K4 cross validation
from sklearn.model_selection import cross_val_score
score = cross_val_score(dt,X_train, y_train,cv = 5, scoring='accuracy')
print("Cross-validation scores:", score*100)
print('mean of scores in decision tree ',np.mean(score)*100)

# training the data
dt.fit(X_train,y_train)
print('Training score',dt.score(X_train,y_train)*100)
```

```python
#prediction
prediction = dt.predict(X_test)
print('test scores',dt.score(X_test,y_test)*100)

#getting Classification report

from sklearn.metrics import classification_report
print(classification_report(prediction,y_test))

#drawing the tree
plt.figure(figsize = (15,8))
from sklearn import tree
tree.plot_tree(dt.fit(X_train, y_train))
```

# MAKING THE DEPTH TO 5

```python
#test and train split
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle=True,test_size  = .30,random_state = 45)
```

```python
#encoding 
x = [i for i in X.columns if df[i].dtypes == 'O' ]
for i in x:
    X_train[i] = encoder.fit_transform(X_train[i])
    X_test[i]= encoder.transform(X_test[i])
    
#depth of tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', max_depth=5,random_state = 1)

#k fold
from sklearn.model_selection import cross_val_score
score = cross_val_score(dt,X_train, y_train,cv = 5, scoring='accuracy')
print("Cross-validation scores:", score*100)
print('mean of scores in decision tree ',np.mean(score)*100)

# training

dt.fit(X_train,y_train)
print('train scores',dt.score(X_train,y_train)*100)
prediction = dt.predict(X_test)
print('test_scores',dt.score(X_test,y_test)*100)
```
